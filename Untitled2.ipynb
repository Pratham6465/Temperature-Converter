```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame=True)
df = iris.frame
```

```python
df['species'] = iris.target_names[iris.target]
```

```python
group_stats = df.groupby('species')['sepal length (cm)'].agg(['mean','median','min','max','std'])
```

```python
print(group_stats)
```

```python
mean_list = group_stats['mean'].tolist()
print("\nList of mean values for each species:", mean_list)
```

```python
setosa = df[df['species'] == 'setosa']
versicolor = df[df['species'] == 'versicolor']
virginica = df[df['species'] == 'virginica']
```

```python
print("\n--- Iris Setosa Statistics ---")
print(setosa.describe())

print("\n--- Iris Versicolor Statistics ---")
print(versicolor.describe())

print("\n--- Iris Virginica Statistics ---")
print(virginica.describe())
```
